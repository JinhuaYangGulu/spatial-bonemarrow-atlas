```python
# This script takes the cluster csvs generated by Export_Cluster_CSVs.R and exports labeled masks for each image from the NBM atlas

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tifffile
```

```python
sample_names =  ["SB67_NBM27_H10_CODEX_Mesmer","SB67_NBM28_H14_CODEX_Mesmer","SB67_NBM31_H32_CODEX_Mesmer",
"SB67_NBM32_H37_CODEX_Mesmer", "SB67_NBM33_H36_CODEX_Mesmer", "SB67_NBM34_H38_CODEX_Mesmer","SB67_NBM36_H26_CODEX_Mesmer",
"SB67_NBM37_H35_CODEX_Mesmer", "SB67_NBM38_H33_CODEX_Mesmer", "SB67_NBM39_H41_CODEX_Mesmer","SB67_NBM40_H39_CODEX_Mesmer",
"SB67_NBM41_H27_CODEX_Mesmer"]
```

```python
for name in sample_names:
	print(name)
	clusters = '/mnt/isilon/tan_lab_imaging/Analysis/bandyopads/NBM_CODEX_Atlas_NeighborhoodMasks/Masks_ForManuscript/NBM_Neighborhood_ClusterCSVs/Filtered_'+name+'_neighborhood10__12312023.csv' 
	mask = '/mnt/isilon/tan_lab_imaging/Analysis/bandyopads/NBM_CODEX_Atlas_NeighborhoodMasks/Masks_ForManuscript/SegmentationMasks/'+name+'_Scan1_segmentation_cell.tif'
	output = '/mnt/isilon/tan_lab_imaging/Analysis/bandyopads/NBM_CODEX_Atlas_NeighborhoodMasks/Masks_ForManuscript/LabeledNBM_Neighborhoods_Mask/Filtered_'+name+'_Nb_LabeledMask.tiff'
	print('Reading Clusters')
	clusters = pd.read_csv(clusters)
	print('Reading Mask')
	mask = tifffile.imread(mask)

	# Initialize numpy array from segmentation mask
	print('Initializing numpy array')
	masks_out = mask.copy()
	masks_out = np.array(masks_out).astype(int)
	print(masks_out.dtype)

	#Convert those cells that have been filtered out to zero
	print('Filtering cells')
	filter_mask = np.isin(masks_out,clusters['CellID'], invert=True)
	masks_out[filter_mask] = 0

	# Create dictionary mapping Cell ID keys to cluster values
	print('Creating mapping')
	mapping = dict(zip(clusters["CellID"], clusters["Annotation"]))

	# Define function to change the values in input array by key-value pairs from dict
	def mp(entry):
	    return mapping[entry] if entry in mapping.keys() else entry

	print('Vectorizing array')
	mp = np.vectorize(mp)

	print('Converting Values') 
	out = mp(masks_out)

	print('Saving mask')
	tifffile.imwrite(output, out)
```
